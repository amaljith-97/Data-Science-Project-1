### Bilirubin: 
It is a red-orange compound that occurs in the normal catabolic pathway that breaks down heme in vertebrates. This catabolism is a necessary process in the body's clearance of waste products that arise from the destruction of aged or abnormal red blood cells #For adults over 18, normal total bilirubin can be up to 1.2 milligrams per deciliter (mg/dl) of blood Direct bilirubin = Conjugated bilirubin + delta bilirubin
### total bilirubin:
<21(micromole/litre), <1.23(milligram/ decilitre) direct bilirubin: 1.0–5.1(micromole/litre) 0–0.3(milligram/ decilitre) An elevation greater than 20 mg/dL suggests severe liver disease. In patients with hepatitis-induced acute liver failure, a serum total bilirubin level > 17.5 mg/dL (300 mmol/L) is a criterion for predicting death and the need for liver transplantation
###  Alkaline phosphatase: 
is secreted by enterocytes, and seems to play a pivotal role in intestinal homeostasis and protection as well as in suppressing inflammation via repression of the downstream Toll-like receptor (TLR)-4-dependent and MyD88-dependent inflammatory cascade The normal range is 44 to 147 international units per liter (IU/L) or 0.73 to 2.45 microkatal per liter (µkat/L) When liver injury occurs, the level of ALP in the circulation increases
### Alamine Aminotransferase: 
On average, the normal levels of SGPT range between 7-56 units/litre of serum. If the SGPT levels in your liver blood test are greater than 56 units/litre, it is referred to as the danger level of SGPT and may indicate any form of injury or damage in the liver ALT levels greater than 15 times the normal range indicate severe acute liver cell injury and evaluation should be initiated immediately
Male 8–40 IU/L Female 6–34 IU/L High levels of AST in the blood may be a sign of hepatitis, cirrhosis, mononucleosis, or other liver diseases
### Total Protiens
The reference range for total protein is typically 60-80g/L. (It is also sometimes reported as "6.0-8.0g/dl") A high total protein level could indicate dehydration or a certain type of cancer, such as multiple myeloma, that causes protein to accumulate abnormally. If the result of a total protein test is abnormal, further tests will be needed to identify which proteins are too high or too low
### Albumin: 
is a protein made by your liver. Albumin enters your bloodstream and helps keep fluid from leaking out of your blood vessels into other tissues. It is also carries hormones, vitamins, and enzymes throughout your body The normal range is 3.4 to 5.4 g/dL (34 to 54 g/L). low level might indicate liver disease
### Albumin and Globulin Ratio:
The normal range for albumin/globulin ratio is over 1 , usually around 1 to 2. That's because there's a bit more albumin than globulin in serum protein. If your body is producing too much or too little of either protein, your A/G ratio will be classified as high or low. A total protein and albumin/globulin (A/G) ratio test measures the total amount of protein in your blood

```python
#outlier handling
fig , ax  = plt.subplots(4, 2, figsize  = (10 ,12))
sns.boxplot(df.Total_Bilirubin, ax  = ax[0,0])
sns.boxplot(df.Direct_Bilirubin, ax  = ax[0,1])
sns.boxplot(df.Alkaline_Phosphotase, ax  = ax[1,0])
sns.boxplot(df.Alamine_Aminotransferase, ax  = ax[1,1])
sns.boxplot(df.Aspartate_Aminotransferase, ax  = ax[2,0])
sns.boxplot(df.Total_Protiens, ax  = ax[2,1])
sns.boxplot(df.Albumin, ax  = ax[3,0])
sns.boxplot(df.Albumin_Globulin_Ratio, ax  = ax[3,1])

```

```python
#outlier treament
# total bilirubin limit is around 1.2
# Alkaline_Phosphotase limit is around 130
# Alamine_Aminotransferase limit is around 7-40
# Aspartate_Aminotransferase limit is around 15-40
# Albumin_Globulin_Ratio limit around 0.8-2.0

def find_ranges(variable):
    
    q1  = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr  = q3  - q1
    lower_range  =  q1  - 1.5 * iqr
    upper_range  =  q3 + 1.5* iqr
    return lower_range , upper_range
```

```python
lower_totbil , upper_totbil =  find_ranges('Total_Bilirubin')
lower_dirbil , upper_dirbil  = find_ranges('Direct_Bilirubin')
lower_alk , upper_alk =  find_ranges('Alkaline_Phosphotase')
lower_alm, upper_alm  =  find_ranges('Alamine_Aminotransferase')
lower_asp ,upper_asp  = find_ranges('Aspartate_Aminotransferase')
lower_alb , upper_alb  = find_ranges('Albumin_Globulin_Ratio')
```

```python
df.Total_Bilirubin =  np.where(df.Total_Bilirubin > upper_totbil , upper_totbil , np.where(
df.Total_Bilirubin < lower_totbil , lower_totbil , df.Total_Bilirubin
))
df.Direct_Bilirubin =  np.where(df.Direct_Bilirubin > upper_dirbil , upper_dirbil , np.where(
df.Direct_Bilirubin < lower_dirbil , lower_dirbil , df.Direct_Bilirubin
))

df.Alkaline_Phosphotase =  np.where(df.Alkaline_Phosphotase > upper_alk , upper_alk, np.where(
df.Alkaline_Phosphotase < lower_alk , lower_alk , df.Alkaline_Phosphotase
))

df.Alamine_Aminotransferase =  np.where(df.Alamine_Aminotransferase > upper_alm , upper_alm , np.where(
df.Alamine_Aminotransferase < lower_alm , lower_alm , df.Alamine_Aminotransferase
))

df.Aspartate_Aminotransferase =  np.where(df.Aspartate_Aminotransferase > upper_asp , upper_asp , np.where(
df.Aspartate_Aminotransferase < lower_asp , lower_asp , df.Aspartate_Aminotransferase
))

df.Albumin_Globulin_Ratio =  np.where(df.Albumin_Globulin_Ratio > upper_alb , upper_alb , np.where(
df.Albumin_Globulin_Ratio < lower_alb , lower_alb , df.Albumin_Globulin_Ratio
))
```
